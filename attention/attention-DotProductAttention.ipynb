缩放点积注意力
点积计算效率较高，但要求 q和k是同维度的，设其维度为d 则：
score(q, k) = q^T k / sqrt(d)
缩放点积注意力不需要训练 wq，wk，wv矩阵了

```python
import torch
import math
from torch import nn
```

```python
class DotProductAttention(nn.Module):
    def __init__(self) -> None:
        super().__init__()

    # queries.shape(len, d)
    # keys.shape(len, len - 1, d)
    # values.shape(len, len - 1, v)
    def forward(self, queries, keys, values):
        d = queries.shape[-1]
        queries =  queries.unsqueeze(-1)
        scores = torch.bmm( keys, queries) /  (d**0.5)
        self.attention_weights = nn.functional.softmax(scores,dim = 1)
        preds = (self.attention_weights * values).sum(dim=1)
        return preds
```

```python
query_size = 4
key_size = 4
value_size = 3
net = DotProductAttention()

def getKeysAndValues(n_train, a_size, a_values):
    a_values = a_values.repeat(n_train,1).reshape(n_train, n_train, a_size)
    values = a_values[(1 - torch.eye(n_train)).type(torch.bool)].reshape(n_train, n_train - 1, a_size)
    return values

n_train = 50
queries = torch.randn(n_train, query_size)
keys = torch.randn(n_train, key_size)
keys = getKeysAndValues(n_train, key_size, keys)


values = torch.randn(n_train, value_size)
values = getKeysAndValues(n_train, value_size, values)

net(queries, keys, values)

```
